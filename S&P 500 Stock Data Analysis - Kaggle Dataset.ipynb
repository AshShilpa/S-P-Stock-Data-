```python
# Merge stock data with company info
df = pd.merge(df1, df2, left_on='Name', right_on='Symbol', how='inner')
```

```python
df.isnull().sum()
df.fillna(0,inplace=True)
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df['date'] = pd.to_datetime(df['date'])
df.sort_values(by=['Symbol', 'date'], inplace=True)
```

```python
print(df.columns)
```

```python
df.rename(columns={
    'Price/Earnings': 'pe_ratio',
    'Earnings/Share': 'eps',
    'Price/Sales': 'price_sales',
    'Price/Book': 'price_book'
}, inplace=True)
```

```python
df.isnull().sum()
df.fillna(value={col: 0 for col in df.select_dtypes(include=[np.number]).columns}, inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# Check and Format Date
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)

# More Formatting
df['Sector'] = df['Sector'].astype('category')
df['Symbol'] = df['Symbol'].str.upper()
df['company'] = df['Name_y'].str.lower()
```

```python
# Transformation
# Ensure 'price' column exists (fallback to 'close' or 'open')
if 'price' not in df.columns:
    if 'close' in df.columns:
        df['price'] = df['close']
    elif 'open' in df.columns:
        df['price'] = df['open']
    else:
        raise KeyError("No column found for 'price', 'close', or 'open'")

df['price_to_sales'] = df['price'] / df['price_sales']
df['price_category'] = df['price'].apply(lambda x: 'High' if x > 100 else 'Low')

available_cols = df.columns
sector_agg = df.groupby('Sector').agg({col: 'mean' for col in ['price', 'market_cap'] if col in available_cols})
pivot_price = df.pivot_table(index='Sector', values='price', aggfunc='mean')
df_reset_temp = df.reset_index() if 'date' in df.index.names else df.copy()
melted_df = df_reset_temp.melt(id_vars=['Sector'], value_vars=['price', 'eps'])
```

```python
# Merge & Concat example
concat_df = pd.concat([df[['price', 'eps']].head(), df[['price', 'eps']].tail()])

# Visualization
sns.histplot(df['price'], bins=30)
plt.title("Price Distribution")
plt.show()

sns.boxplot(x='Sector', y='price', data=df)
plt.xticks(rotation=90)
plt.title("Price by Sector")
plt.show()

sns.scatterplot(x='eps', y='price', hue='Sector', data=df)
plt.title("EPS vs Price")
plt.show()
```

```python
# Bar chart: Average Price by Sector
avg_price_sector = df.groupby('Sector')['price'].mean().sort_values()
avg_price_sector.plot(kind='barh', title='Average Price by Sector')
plt.xlabel('Average Price')
plt.ylabel('Sector')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap for numeric fields
plt.figure(figsize=(10, 6))
numeric_cols = [col for col in ['price', 'eps', 'market_cap', 'pe_ratio', 'price_sales', 'price_book'] if col in df.columns]
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# KDE plot for price
sns.kdeplot(df['price'], shade=True)
plt.title("Price Density Estimate")
plt.show()
```

```python
# Strip plot to show price distribution by sector
plt.figure(figsize=(12, 6))
sns.stripplot(x='Sector', y='price', data=df, jitter=True, alpha=0.6)
plt.xticks(rotation=90)
plt.title("Price Distribution per Sector")
plt.show()
```

```python
# Rolling average price
rolling_avg = df['price'].rolling(window=30).mean()
df['rolling_avg'] = rolling_avg
```

```python
# Rolling Standard Deviation (Volatility)
df['rolling_std'] = df['price'].rolling(window=30).std()

# Plot for a sample company
sample_symbol = 'AAPL'  # Replace with valid symbol
if sample_symbol in df['Symbol'].unique():
    df_symbol = df[df['Symbol'] == sample_symbol]
    df_symbol[['price', 'rolling_avg', 'rolling_std']].plot(title=f"{sample_symbol} - Price, Rolling Avg & Volatility")
    plt.ylabel("Price")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if df.index.name == 'date':
    monthly_avg = df['price'].resample('M').mean()
    monthly_avg.plot(title='Monthly Avg Price')
    plt.ylabel('Price')
    plt.show()
```

```python
#Top N Companies by EPS
top_eps = df.groupby('Symbol')['eps'].mean().sort_values(ascending=False).head(10)
top_eps.plot(kind='bar', title='Top 10 Companies by EPS')
plt.ylabel('EPS')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Regression Plot: EPS vs. Price
sns.lmplot(data=df, x='eps', y='price', scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'})
plt.title("Linear Regression: EPS vs. Price")
plt.show()
```

```python
#Boxplot of EPS by Sector
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sector', y='eps', data=df)
plt.xticks(rotation=90)
plt.title("EPS Distribution by Sector")
plt.show()
```

```python
high_eps = df.query('eps > 5')
selected_sectors = df[df['Sector'].isin(['Technology', 'Healthcare'])]

# Statistical Ops
stats_cols = [col for col in ['price', 'eps', 'market_cap'] if col in df.columns]
print("Correlation:", df[stats_cols].corr())
print("Kurtosis:", kurtosis(df['price']))
print("Skew:", skew(df['price']))

# Normalization & Standardization
df['price_zscore'] = (df['price'] - df['price'].mean()) / df['price'].std()
df['price_norm'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
```

```python
df_reset = df.reset_index()
# Drop duplicates for pivot compatibility
pivot_df = df_reset.drop_duplicates(subset=['Symbol', 'Sector'])
pivoted = pivot_df.pivot(index='Symbol', columns='Sector', values='price')
```
